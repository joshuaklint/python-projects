```python
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
```

```python
import cufflinks as cf
```

```python
py.offline.init_notebook_mode(connected = True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,3), columns = ['A','B','C'])
df['A']  = df['A'].cumsum()+20
df['B']  = df['B'].cumsum()+20
df['C']  = df['C'].cumsum()+20
```

```python
df.shape
```

```python
plt.plot(df)
```

```python
cf.getThemes()
```

```python
cf.set_config_file(theme = 'henanigans')
```

```python
df.iplot(x = 'A', y = 'B', mode= 'markers', size = 14)
```
