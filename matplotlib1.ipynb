```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,20)
```

```python
y = np.random.randint(1,20,20)
```

```python
plt.plot(y)
plt.plot(x,y)
plt.ylim(0,20)
plt.xlim(0,20)
plt.xticks([1,5,10,15,20])
```

```python
y = np.sort(y)
```

```python
plt.subplot(121)
plt.plot(y,'gD-',markersize = 2)
plt.subplot(122)
plt.plot(x,y,'c*-', markersize = 2)
plt.xlabel(' Linspace Random values', color = 'green')
plt.ylabel('Random valus',color = 'blue')
plt.title('Random plot', color = 'black')
```

```python
plt.subplot(211)
plt.plot(y)
plt.subplot(212)
plt.plot(x,y,'c--')
plt.xlabel(' Linspace Random values', color = 'green')
plt.ylabel('Random valus',color = 'blue')
plt.title('Random plot', color = 'black')
```

```python
plt.plot(x,y,'c--')
plt.xlabel(' Linspace Random values', color = 'green')
plt.ylabel('Random valus',color = 'blue')
plt.title('Random plot', color = 'black')
```

```python
y2 = y*x
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axe1  = fig.add_axes([.1,.6,.4,.3])

axes.plot(x,y,'g--')
axes.set_xlabel('X- values', color = 'cyan')
axes.set_ylabel('Y- values', color = 'cyan')
axes.set_title('y-square plot', color = 'cyan')
axes.legend('random plot', loc = 1)
axe1.plot(x,y2, 'c*-')
axe1.set_xlabel('X- values', color = 'blue')
axe1.set_ylabel('Y- values',color = 'blue')
axe1.set_title('y plot',color = 'blue')
```

```python
fig = plt.figure(figsize = [10,4])
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'c--')
ax.set_xlabel(' Linspace Random values', color = 'green')
ax.set_ylabel('Random valus',color = 'blue')
ax.set_title('Random plot', color = 'black')
```
