```python
def sinplot(y = 1):
    x = np.linspace(0,14,100)
    for k in range(1,7):
        plt.plot(x, np.sin(x+k*0.5)*(7-k)*y)
```

```python
sinplot(2)
```

```python
sns.set_style('white')
sns.set_context('talk', font_scale = 2)
sinplot()
```
